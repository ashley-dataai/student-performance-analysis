# Student Performance Analysis

This notebook demonstrates basic data cleaning and visualization using Python and Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Attendance (%)': [90, 80, 75, 95, 60],
    'Grade': [88, 76, 79, 92, 70]
}
df = pd.DataFrame(data)
df
```

```python
# Basic statistics
df.describe()
```

```python
# Visualization
plt.figure(figsize=(8, 5))
plt.bar(df['Student'], df['Grade'], color='skyblue')
plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```
